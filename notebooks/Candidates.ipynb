```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_excel('~/Data/Smartvote/election23/23_ch_nr_candidates_de_2024_03_06.xlsx', index_col=0)
parties = dataset.groupby("party_short").ID_candidate.count().sort_values()
parties
```

```python
party_colors = {
    'EVP': '#FFA500',      # Orange
    'Grüne': '#008000',    # Green
    'FDP': '#0000FF',      # Blue
    'SVP': '#4F7942',      # Slate Gray
    'SP': '#FF0000',       # Red
    'GLP': '#32CD32',      # Lime green
    'Die Mitte': '#C56C46' # Bronze
}
```

```python
candidates = dataset.loc[dataset['party_short'].isin(party_colors.keys()) & (dataset['canton'] == 1)]
candidates
```

```python
# Use a regex pattern to filter columns
pattern = r'^answer_\d+$'  # ^ and $ ensure no characters before or after the pattern
reactions = candidates.filter(regex=pattern).dropna()/100
reactions
```

```python
reactions.to_csv('../src/data/csv/reactions.csv')
```

```python
candidates = candidates.loc[reactions.index].filter(['party_short'])
candidates['color'] = candidates.party_short.map(party_colors)
candidates
```
